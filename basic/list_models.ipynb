# Basic GPT 3.5 Turbo API call

```python
import openai
from environs import Env
import pandas as pd
from IPython.display import JSON

env = Env()

with env.prefixed("AZURE_"):
    openai.api_key = env("OPENAI_API_KEY")
    openai.api_type = env("OPENAI_API_TYPE", "open_ai")
    openai.api_base = env("OPENAI_API_BASE", "https://api.openai.com/v1")
    openai.api_version = env("OPENAI_API_VERSION", None)
```

# Get models in JSON

```python
models = openai.Model.list()
```

# Display models in a Pandas Dataframe

```python
df = pd.DataFrame.from_dict(models["data"]) # load JSON into DataFrames
df = df[['id', 'lifecycle_status', 'status', 'created_at', 'updated_at']] # Select columns we want
df = df.astype({'created_at': 'datetime64[s]', 'updated_at': 'datetime64[s]'}) # Set column data types
df
```
